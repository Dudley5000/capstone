```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
from datetime import datetime
dateparse = lambda x: datetime.strptime(x, '%Y-%m-%d')
```

```python
dfPrices = pd.read_csv('stock_prices.csv',index_col=0,parse_dates=['Date'],date_parser=dateparse)
dfPrices.head()
```

```python
dfKeyStats = pd.read_csv('keystats.csv',index_col=0)
dfKeyStats.sort_values(by=['Ticker','Date'], inplace=True, ascending=True)
dfKeyStats.head()
```

```python
# Profitability is defined as the condition where a stock's percentage change is greater
# than that of the S&P (plus an optional offset percentage)
isProfitable = lambda stock, sp500, gap=0 : stock > (sp500 + gap)
```

```python
# True/False ratio of stocks performing at least 50 percent better than the S&P
dfKeyStats['isProfitable'] = dfKeyStats.apply(lambda x: isProfitable(x['stock_p_change'], x['SP500_p_change'], 50), axis=1)
print((dfKeyStats[dfKeyStats['isProfitable'] == True].size, dfKeyStats[dfKeyStats['isProfitable'] == False].size))

# True/False ratio of stocks performing at least 400 percent better than the S&P
dfKeyStats['isProfitable'] = dfKeyStats.apply(lambda x: isProfitable(x['stock_p_change'], x['SP500_p_change'], 400), axis=1)
print((dfKeyStats[dfKeyStats['isProfitable'] == True].size, dfKeyStats[dfKeyStats['isProfitable'] == False].size))
```

```python
# New dataframe to track and display the range of ratios shown above
myDict = {}

for i in range(0, 200, 5):
    dfKeyStats['isProfitable'] = dfKeyStats.apply(lambda x: isProfitable(x['stock_p_change'], x['SP500_p_change'], i), axis=1)
    
    ratio = dfKeyStats[dfKeyStats['isProfitable'] == True].size / dfKeyStats.size
    myDict[i] = ratio * 100
    
import pprint
pp = pprint.PrettyPrinter(indent=4)    
pp.pprint(myDict)
```

```python
df = pd.DataFrame.from_dict(data=myDict, orient='index')
```

```python
plt.figure(figsize=(12, 5))
plt.axis([0,150,0,60])
plt.xlabel('Percentage improvement over S&P')
plt.ylabel('Percentage of records')
plt.plot(df)
```

```python
corr_matrix = dfKeyStats.corr()

mask = np.zeros_like(corr_matrix, dtype=np.bool)
# np.zeros_like() returns an array of zeros with the same shape and type as the given array. 
# By passing in the correlation matrix, we get an array of zeros. The dtype=np.bool parameter 
# overrides the data type, so our array is an array of booleans.

mask[np.triu_indices_from(mask)]= True
# np.triu_indices_from(mask) returns the indices for the upper triangle of the array. 
# This yields a mask that we can use to generate the heatmap

f, ax = plt.subplots(figsize=(16, 15))

heatmap = sns.heatmap(corr_matrix,
                      mask = mask,
                      square = True,
                      linewidths = .5,
                      cmap = 'coolwarm',
                      cbar_kws = {'shrink': .4,
                                'ticks' : [-1, -.5, 0, 0.5, 1]},
                      vmin = -1,
                      vmax = 1,
                      annot_kws = {'size': 12})

sns.set_style({'xtick.bottom': True})
```

## K Means Clustering

```python
tickers = list(set(dfKeyStats['Ticker'].tolist()))
```

```python
import yfinance as yf
data = yf.download(tickers=tickers[:100], period = "1y")
```

```python
data['Volume']['NYX']
```

```python
filtered_data = data.dropna(how="any", axis=1)
filtered_data
```

```python
filtered_data['Adj Close'].columns
```

```python
tickers = filtered_data['Close'].columns.tolist()
stock_close = filtered_data['Close']
stock_open = filtered_data['Open']

# Calculate daily stock movement
stock_close = np.array(stock_close).T
stock_open = np.array(stock_open).T

row, col = stock_close.shape

# create movements dataset filled with 0's
movements = np.zeros([row, col])

for i in range(0, row):
 movements[i,:] = np.subtract(stock_close[i,:], stock_open[i,:])

# for i in range(0, len(tickers)):
#  print('Ticker: {}, Change: {}'.format(tickers[i], sum(movements[i][:])))
```

```python
plt.figure(figsize=(18,16))
ax1 = plt.subplot(221)
plt.plot(movements[1][:])
plt.title(tickers[1])

plt.subplot(222, sharey=ax1)
plt.plot(movements[5][:])
plt.title(tickers[5])
plt.show()
```

```python
from sklearn.preprocessing import Normalizer
normalizer = Normalizer()

new = normalizer.fit_transform(movements)

print(new.max())
print(new.min())
print(new.mean())
```

```python
plt.figure(figsize=(18,16))
ax1 = plt.subplot(221)
plt.plot(new[1][:])
plt.title(tickers[1])

plt.subplot(222, sharey=ax1)
plt.plot(new[5][:])
plt.title(tickers[5])
plt.show()
```

```python
from sklearn.pipeline import make_pipeline
from sklearn.cluster import KMeans

CLUSTERS=20
normalizer = Normalizer()
kmeans = KMeans(n_clusters=CLUSTERS, max_iter=1000)
pipeline = make_pipeline(normalizer,kmeans)
```

```python
pipeline.fit(movements)
```

```python
print(kmeans.inertia_)
```

```python
# predict cluster labels
labels = pipeline.predict(movements)

# create a DataFrame aligning labels & companies
df = pd.DataFrame({'labels': labels, 'companies': tickers})

# display df sorted by cluster labels
pd.set_option('display.max_rows', None)
# pd.reset_option('display.max_rows')

df.sort_values('labels')
```

```python
# PCA
from sklearn.decomposition import PCA 

# visualize the results
reduced_data = PCA(n_components = 2).fit_transform(new)

# run kmeans on reduced data
kmeans = KMeans(n_clusters=CLUSTERS)
kmeans.fit(reduced_data)
labels = kmeans.predict(reduced_data)

# create DataFrame aligning labels & companies
df = pd.DataFrame({'labels': labels, 'companies': tickers})

# Display df sorted by cluster labels
print(df.sort_values('labels'))
```

```python
# Define step size of mesh
h = 0.01

# plot the decision boundary
x_min, x_max = reduced_data[:, 0].min() - 1, reduced_data[:,0].max() + 1
y_min, y_max = reduced_data[:, 1].min() - 1, reduced_data[:,1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))

# Obtain labels for each point in the mesh using our trained model
Z = kmeans.predict(np.c_[xx.ravel(), yy.ravel()])

# Put the result into a color plot
Z = Z.reshape(xx.shape)

# define colorplot
cmap = plt.cm.Paired

# plot figure
plt.clf()
plt.figure(figsize=(10,10))
plt.imshow(Z, interpolation='nearest',
 extent = (xx.min(), xx.max(), yy.min(), yy.max()),
 cmap = cmap,
 aspect = 'auto', origin='lower')
plt.plot(reduced_data[:, 0], reduced_data[:, 1], 'k.', markersize=5)


# plot the centroid of each cluster as a white X
centroids = kmeans.cluster_centers_
plt.scatter(centroids[:, 0], centroids[:, 1],
 marker='x', s=169, linewidth=3,
 color='w', zorder=10)

plt.title('K-Means Clustering on Stock Market Movements (PCA-Reduced Data)')
plt.xlim(x_min, x_max)
plt.ylim(y_min, y_max)
plt.show()
```
